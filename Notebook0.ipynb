## 2.4 Variables

At this point, if we want to perform different operations on the same expression, we have to copy-paste the expression for each operation.

````python
print(5 - 4)
print((5 - 4)/4)
print(((5 - 4) / 4) * 100)
````

This becomes troublesome to manage very quickly. Luckily, in Python, we have **variables** that help make this easier. Variables are names that can be assigned to expressions via *assignment statements*. To assign a variable to an expression, put the variable name on the left side of an ``=`` sign and an expression on the right.

````python
ten = 3 * 2 + 4
````

When you run that cell, Python first computes the value of the expression on the right-hand side, `3 * 2 + 4`, which is the number 10.  Then it assigns that value to the name `ten`.  At that point, the code in the cell is done running.

After you run that cell, the value 10 is bound to the name `ten`:

````python
ten
````

Now, you can perform any operation on the variable ``ten`` as you would with the expression ``3 * 2 + 4``.

````python
ten + 1
````

````python
ten / 2
````

Note that, as you could have assigned any other expression to the variable name ``ten``, you can easily override its meaning if you perform another assignment in later operations.

````python
ten = 3 * 2 + 4
print(ten)
ten = 3 * 2 + 5
print(ten)
````

This is often undesirable as you can lose track of what the variable name is actually assigned to. Therefore, you should avoid assigning the same variable name to different expressions, unless the override is intentional.

Here are some rules and conventions when it comes to creating variable names.


1. You cannot have any space in a variable name.
```python
number ten = 3 * 2 + 4
```

2. If you want to use multiple words for a variable name, connect them via an underscore ``_``. (This is called [snake_case](https://en.wikipedia.org/wiki/Snake_case))
```python
number_ten = 3 * 2 + 4
```

3. Sometimes you will also see multi-word variable names in the format of [camelCase](https://en.wikipedia.org/wiki/Camel_case). Try to not use this in your Python code, however, as it goes against usual standards for variable names.
```python
numberTen = 3 * 2 + 4
```

4. You cannot start a variable name with a number (but you can end with one).

```python
10number = 3 * 2 + 4 # This will produce an error

number10 = 3 * 2 + 4 # This is fine
```

5. Try to make your variable names descriptive so that you do not forget what the value stored in that variable is supposed to represent!

```python
x = 5 # Not a great name

number_of_beans_in_jar = 5 # Better since now I won't mix this up with other variables
```

## 2.6 Booleans

People sometimes talk about a computer as "a bunch of ones and zeros", so it is natural for any programming language to have a data type specifically for that. The Boolean data type represents only two values ``True`` (``1``) or ``False`` (``0``). Note that capitalization matters here.

Boolean values allows for lots of logical expressions and operations. The simplest of them are ``>``, ``>=``, ``==``, ``<``, and ``<=``. You can use these to compare different values.

> ⚠️ WARNING ⚠️ the `==` operator is used to check for equality, and the `=` operator is used for variable assignment. Don't mix them up!!

````python
print(2 < 3)
print(5 > 7)
````

You can also compare outputs of math operations. Use parentheses to specify order of operations. In the following expression, Python first computes ``5/2`` and ``10/4``, then assesses whether the results are equal.

Another common operation is ``and``, ``or``, and ``not``. For example, ``and`` evaluates to ``True`` if both expressions it assesses are ``True`` or evaluate to ``True``.

````python
# "and" evaluates to True if both expressions are True
print("and ============================")
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print("")

# "or" evaluates to True if at least one of the expressions is True
print("or ============================")
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print("")

# "not" evaluates to True if the expression is False
print("not ============================")
print(not True)
print(not False)
````

## 2.12 Advanced Operations

Python is a reasonably flexible language, so sometimes you will encounter operations that go beyond what we have discussed above. Here are some other operations you can perform to code more efficiently.

### Formatted String

Consider a numerical variable ``score``. We want to print out the sentence ``"The score is x."``, with ``x`` replaced by the score. We can do this by first converting the number to a string using an operation called ``str()``, then use concatenation.

````python
score = 75
score_str = str(score)
sentence = "The score is " + score_str + "."
print(sentence)
````

Or we can do this in one line.

````python
print("The score is " + str(score) + ".")
````

However, this is not very efficient. A better way to do this is to use **formatted strings** (f-strings).

````python
print(f"The score is {score}.")
````

We place the letter f in front of the starting quotation mark, and put the variable, wrapped in curly brackets, in the part of the sentence we would like it to replace. We can also place an expression in the curly brackets, and Python will execute the expression before adding it to the string.

````python
print(f"The score is greater than 70: {score > 70}")
````

### Conditional Statments

Consider a variable ``score``. We want to print out the corresponding letter grade. Currently, this is the best we can do:

````python
score = 75

print(f"The grade is A: {score >= 90}")
print(f"The grade is B: {score >= 80}")
print(f"The grade is C: {score >= 70}")
print(f"The grade is F: {score < 70}")
````

A more simple way is to use **if-else** statements.

````python
if score >= 90:
    print("The grade is A.")
elif score >= 80:
    print("The grade is B.")
elif score >= 70:
    print("The grade is C.")
else:
    print("The grade is F.")
````

In natural language, the code does the following.

- First assess if the ``score`` is at least 90, if so, print ``"The grade is A"``.
- If the first assessment is not true, assess whether ``score`` is at least 80. If so, print ``"The grade is B"``.
- If the first and second assessments are not true, assess whether ``score`` is at least 70. If so, print ``"The grade is C"``.
- If none of the above is true, print ``"The grade is F"``.

Additionally, note the **indentation**: Every time we have a conditional statement, we need to indent the code we would like Python to execute if the statement is True.

### For-Loops

Consider a list of numbers called ``numbers``. If we would like to print out each element of the list, raised to the power of 3. We can do the following.

````python
numbers = [1.21, 10.75, 3.45, 5.99]

print(numbers[0] ** 3)
print(numbers[1] ** 3)
print(numbers[2] ** 3)
print(numbers[3] ** 3)
````

This is inefficient (consider if the list has 1000 numbers). Instead, Python has a special operation called ``for-loop`` that handles repetitive operations like this.

````python
for x in numbers:
    print(x**3)
````

Note we achieve the exact same result, but in 2 lines instead of 4. In natural language, the code does the following.

- Look at the list ``numbers``.
- Loop 1: Assign the variable name ``x`` to the first element of ``numbers`` (equavalent to ``x=numbers[0]``), then perform the specified operation (i.e. ``print(x**3)``).
- Loop 2: Assign the variable name ``x`` to the second element of ``numbers`` (equavalent to ``x=numbers[1]``), then perform the specified operation (i.e. ``print(x**3)``).
- Loop 3: Assign the variable name ``x`` to the third element of ``numbers`` (equavalent to ``x=numbers[2]``), then perform the specified operation (i.e. ``print(x**3)``).
- Loop 4: Assign the variable name ``x`` to the third element of ``numbers`` (equavalent to ``x=numbers[3]``), then perform the specified operation (i.e. ``print(x**3)``).

Notice every last part of the sentence in the last four bullets are exactly the same. Through assigning the same variable to different expressions, a for-loop handles the repetitive task of printing out each number, raised to the power of three.

**Note** Here ``x`` is simply a choice of a placeholder, as we can use any other variable name.

````python
for this_is_a_very_long_variable_name  in numbers:
    print(this_is_a_very_long_variable_name**3)
````

````python
for _ in numbers:
    print(_**3)
````
